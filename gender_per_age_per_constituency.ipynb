### Let us now narrow down to gender statistics as below
### We now pull total number of male and female per county.

```python
county = data.groupby('COUNTY').sum()
county['Ratio of Female to Male'] = county['TOTAL FEMALE']/county['TOTAL MALE']
county = county[['TOTAL FEMALE', 'TOTAL MALE', 'Ratio of Female to Male']]
#county
county.plot(kind='bar', figsize=(14,7), y=['TOTAL FEMALE','TOTAL MALE'])
plt.title("Total Gender Relations per County", fontsize=22, fontweight='bold')
plt.xlabel("County", fontsize=18, fontweight='bold')
plt.ylabel("Number of voters", fontsize=18, fontweight='bold')
```

#### We can see that in all the counties the male voters are the majority.

## Voter Aggregation as below:
### 1. Counties where the number of female voters is greater than national average

```python
countybig =county.where(county['TOTAL FEMALE'] > county['TOTAL FEMALE'].mean()).dropna()
#countybig.dropna() #moved to the end of the above line
#county['TOTAL FEMALE'].min()
#county['TOTAL FEMALE'].max()
#county['TOTAL FEMALE'].mean()
countybig.plot(kind='bar', figsize=(14,7), y=['TOTAL FEMALE','TOTAL MALE'])
plt.title("Counties where Number of Femal voters is Greater than National Average", fontsize=22, fontweight='bold')
plt.xlabel("County", fontsize=18, fontweight='bold')
plt.ylabel("Number of voters", fontsize=18, fontweight='bold')
```

### 2. Counties where the number of female voters is less than national average

```python
countybig =county.where(county['TOTAL FEMALE'] < county['TOTAL FEMALE'].mean()).dropna()
#countybig.dropna() #moved to the end of the above line
countybig.plot(kind='bar', stacked=True, figsize=(14,7), y=['TOTAL FEMALE','TOTAL MALE'])
plt.title("Counties where Number of Female voters is Less than National Average", fontsize=22, fontweight='bold')
plt.xlabel("County", fontsize=18, fontweight='bold')
plt.ylabel("Number of voters", fontsize=18, fontweight='bold')
```

### Counties With the Lowest Number of Young Female Voters

```python
countyAgeGroupLowest = data.groupby('COUNTY').sum()
countyAgeGroupLowest = data.sort_values(by='18 to 24 YRS(F)', ascending = True)
countyAgeGroupLowest[['COUNTY','CONSTITUENCY', '18 to 24 YRS(F)']].head(5)
```

### Constituencies With the Highest Number of Young Female Voters

```python
countyAgeGroupHighest = data.sort_values(by='18 to 24 YRS(F)', ascending = False)
countyAgeGroupHighestPie = countyAgeGroupHighest[['COUNTY','CONSTITUENCY', '18 to 24 YRS(F)']].head(5)
labels = countyAgeGroupHighestPie['CONSTITUENCY']
countyAgeGroupHighestPie.plot(kind='pie', y='18 to 24 YRS(F)', radius=2.5, labels=labels)
```

### Counties whith highest number of old male voters above 55 Years 

```python
oldpeople = data.sort_values(by='OVER 55 YRS(M)', ascending=False)
oldpeoplepie = oldpeople[['COUNTY', 'CONSTITUENCY', 'OVER 55 YRS(M)']].head()
labels = oldpeoplepie['COUNTY']
oldpeoplepie.plot(kind='pie', y='OVER 55 YRS(M)', radius=2.5, labels=labels)
```
