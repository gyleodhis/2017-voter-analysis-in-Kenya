### Let us now narrow down to gender statistics as below
### We now pull total number of male and female per county.

```python
county = data.groupby('COUNTY').sum()
county['Ratio of Female to Male'] = county['TOTAL FEMALE']/county['TOTAL MALE']
county = county[['TOTAL FEMALE', 'TOTAL MALE', 'Ratio of Female to Male']]
county
```

#### We can see that in all the counties the male voters are the majority.

## Voter Aggregation as below:
### 1. Counties where the number of female voters is greater than national average

```python
countybig =county.where(county['TOTAL FEMALE'] > county['TOTAL FEMALE'].mean())
countybig.dropna()
#county['TOTAL FEMALE'].min()
#county['TOTAL FEMALE'].max()
#county['TOTAL FEMALE'].mean()
```

### 2. Counties where the number of female voters is less than national average

```python
countybig =county.where(county['TOTAL FEMALE'] < county['TOTAL FEMALE'].mean())
countybig.dropna()
```

### Counties With the Lowest Number of Young Female Voters

```python
countyAgeGroupLowest = data.groupby('COUNTY').sum()
countyAgeGroupLowest = data.sort_values(by='18 to 24 YRS(F)', ascending = True)
countyAgeGroupLowest[['COUNTY','CONSTITUENCY', '18 to 24 YRS(F)']].head(5)
```

### Counties With the Highest Number of Young Female Voters

```python
countyAgeGroupHighest = data.sort_values(by='18 to 24 YRS(F)', ascending = False)
countyAgeGroupHighest[['COUNTY','CONSTITUENCY', '18 to 24 YRS(F)']].head(5)
```

### Counties whith highest old male voters above 55 Years 

```python
oldpeople = data.sort_values(by='OVER 55 YRS(M)', ascending=False)
oldpeople[['COUNTY', 'CONSTITUENCY', 'OVER 55 YRS(M)']].head()
```
