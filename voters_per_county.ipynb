```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
data = pd.read_excel('./static/Voters_per_county.xlsm')
#data.head(50)
fig = plt.figure()
data.plot(kind='bar', figsize=(14,7), x='COUNTY_NAME', y=['VOTERS','NO. OF POLLING STATIONS'])
plt.title("Voters Per County", fontsize=22, fontweight='bold')
```

## Average number of voters per polling station in ascending order

```python
#data.groupby("COUNTY_NAME").sum()
#data.insert(4, "No of PEOPLE PER  POLLING STATION", data['VOTERS']/data['NO. OF POLLING STATIONS'])
data['No of PEOPLE PER  POLLING STATION'] = data['VOTERS']/data['NO. OF POLLING STATIONS']
data = data.sort_values(by = "No of PEOPLE PER  POLLING STATION", ascending=True)
data[['COUNTY_NAME','No of PEOPLE PER  POLLING STATION']]
data.plot(kind='bar', figsize=(14,7), x='COUNTY_NAME', y='No of PEOPLE PER  POLLING STATION')
plt.title("Number of People  per Polling Station per County", fontsize=22, fontweight='bold')
```

__Nairobi County__ Has the highest number of registerd voters with __666.327117__ per polling station while __West Pokot County__ registered the lowest numbers with __253.134831__ persons per polling station.

### Which Counties have highest number of voters

```python
#data.groupby("COUNTY_NAME").sum()
countyHighest = data[['COUNTY_NAME','VOTERS']]
voterHighest = countyHighest.sort_values(by = "VOTERS", ascending=False).head(10)
voterHighest.plot(kind='bar', figsize=(14,7), x='COUNTY_NAME', y='VOTERS')
plt.title("Counties with Highest Voters", fontsize=22, fontweight='bold')
```

### Note the Nan Value Stands for diaspora

### Which Counties have lowest number of voters
It is thus clear that Kenyans in __diaspora__ with __19601502__ are many than any number of registered voters in any county followed by __Nairobi.__

```python
voterLowerst = countyHighest.sort_values(by = "VOTERS", ascending=False).tail(10)
voterLowerst.plot(kind='bar', figsize=(14,7), x='COUNTY_NAME', y='VOTERS')
plt.title("Counties with Lowest Voters", fontsize=22, fontweight='bold')
```

__Lamu county__ registerd the least number of voters. Only __69776__ persons where eligible to vote.
