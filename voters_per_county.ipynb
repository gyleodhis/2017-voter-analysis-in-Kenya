```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
data = pd.read_excel('./static/Voters_per_county.xlsm')
data.head(50)
```

## Average number of voters per polling station in ascending order

```python
#data.groupby("COUNTY_NAME").sum()
#data.insert(4, "No of PEOPLE PER  POLLING STATION", data['VOTERS']/data['NO. OF POLLING STATIONS'])
data['No of PEOPLE PER  POLLING STATION'] = data['VOTERS']/data['NO. OF POLLING STATIONS']
data = data.sort_values(by = "No of PEOPLE PER  POLLING STATION", ascending=True)
data[['COUNTY_NAME','No of PEOPLE PER  POLLING STATION']]
```

__Nairobi County__ Has the highest number of registerd voters with __666.327117__ per polling station while __West Pokot County__ registered the lowest numbers with __253.134831__ persons per polling station.
